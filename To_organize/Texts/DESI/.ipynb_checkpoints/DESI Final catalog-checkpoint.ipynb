```python
import os
import sys
import glob
import numpy as np
import  scipy.optimize    as op
import matplotlib.pyplot as plt
import urllib.request
from astropy.io import fits as fits
from astropy.table import Table


file="catQSO_dark_bal_mask_sv13_wLN.fits"
local_dir="/home/hquintero/DESI/"
local_file = os.path.join(local_dir,file)
main = Table.read(local_file)
```

```python
print(f"Totales: {len(main)}")
print(f"Unicos: {len(np.unique(main['TARGETID']))}")
u, i = np.unique(np.array(main['TARGETID']), return_inverse=True)
RE = u[np.bincount(i) > 1]
print(f"Repetidos: {len(RE)}")
```

```python
check = 0

for x in range(len(RE)):
    
    a = (main['TARGETID'] == RE[x])
    s = main[a] #selec
    
    if(len(s)>2):
        check = check + 1
        
    clas = np.array([[s['TARGETID'][0],s['TARGETID'][1]],[s['COADD_EXPTIME'][0],s['COADD_EXPTIME'][1]]])
    
    if ((s['COADD_EXPTIME'][0])>(s['COADD_EXPTIME'][1])):
        
        b = (main['TARGETID'] == s['TARGETID'][1]) & (main['COADD_EXPTIME'] == s['COADD_EXPTIME'][1])
        
    elif ((s['COADD_EXPTIME'][1])>(s['COADD_EXPTIME'][0])):
        
        b = (main['TARGETID'] == s['TARGETID'][0]) & (main['COADD_EXPTIME'] == s['COADD_EXPTIME'][0])
        
    idx = np.where(b)
    
    main.remove_rows(idx)
    
    
    
print(check)

```

```python
print(f"Totales: {len(main)}")
print(f"Unicos: {len(np.unique(main['TARGETID']))}")
u, i = np.unique(np.array(main['TARGETID']), return_inverse=True)
RE = u[np.bincount(i) > 1]
print(f"Repetidos: {len(RE)}")
```

```python
hdu = fits.BinTableHDU(main)
with fits.open('DESIedr_FINALcat.fits', mode='update') as passer:
    passer[1] = hdu 
    passer.flush()
```

### Removing BALs

```python
final_cat = Table.read('DESIedr_FINALcat.fits')
bal_rem = final_cat['BI_CIV']<=0
new_cat = final_cat[bal_rem]
```

```python
hdu = fits.BinTableHDU(new_cat)
with fits.open('DESIedr_Fincat_SinBAL.fits', mode='update') as passer:
    passer[1] = hdu 
    passer.flush()
```
