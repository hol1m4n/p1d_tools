```python
a = (dr14['Z']>=2.8) & (dr14['Z']<=3.1) & (dr14['PLATE']==7339) & (dr14['MJD']==56799) & (dr14['FIBERID']==379)
sep_eboss = dr14[a]
b = (DESI['Z']>=2.8) & (DESI['Z']<=3.1)
sep_DESI = DESI[b]

print(len(sep_eboss),len(sep_DESI))
```

```python
ra1 = np.array(sep_eboss['RA'])
dec1 = np.array(sep_eboss['DEC'])
ra2 = np.array(sep_DESI['TARGET_RA'])
dec2 = np.array(sep_DESI['TARGET_DEC'])

c_eboss = SkyCoord(ra=ra1*u.degree, dec=dec1*u.degree)
c_DESI = SkyCoord(ra=ra2*u.degree, dec=dec2*u.degree)

idx, d2d, d3d = c_eboss.match_to_catalog_sky(c_DESI)
```

```python
ct = Table()

ct['ID'] = np.array(sep_eboss['THING_ID'])
ct['ra'] = np.array(sep_eboss['RA'])
ct['dec'] = np.array(sep_eboss['DEC'])
ct['z'] = np.array(sep_eboss['Z'])
ct['plate'] = np.array(sep_eboss['PLATE'])
ct['mjd'] = np.array(sep_eboss['MJD'])
ct['fiber'] = np.array(sep_eboss['FIBERID'])
ct['idx'] = np.array(idx)
ct['ang_sep'] = np.array(d2d)

clas = ct['ang_sep'] < 0.00001 

ctn = ct[clas]

df = ctn.to_pandas()

pd.set_option('display.max_rows', 137)

df
```
